**Who are the top 10 wealthiest billionaires across all years combined?**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset=dataset.rename(columns={'Net worth\xa0(USD)\n(billion)':'Networth'})
top10=dataset.sort_values('Networth',ascending=False).head(25)
sns.barplot(x='Name',y='Networth',data=top10)
plt.show()
```

```python
# Step 2: Basic Statistical Analysis with NumPy

mean_networth = np.mean(top10['Networth'])
median_networth = np.median(top10['Networth'])
std_networth = np.std(top10['Networth'])
print("\nStatistical Summary of Top 10 Net Worth:")
print(f"Mean: {mean_networth:.2f} billion USD")
print(f"Median: {median_networth:.2f} billion USD")
print(f"Standard Deviation: {std_networth:.2f} billion USD")


if 'Age' in dataset.columns:
    avg_age = np.mean(top10['Age'].dropna())
    print(f"Average Age: {avg_age:.2f} years")

#
# Count of top 10 by Nationality
nationality_count = top10['Nationality'].value_counts()
print("\nNationality Distribution in Top 10:")
print(nationality_count)

# Common sources of wealth
sources_count = top10['Source(s) of wealth'].value_counts()
print("\nSource of Wealth Distribution in Top 10:")
print(sources_count)
# Step 4: Visualizations with Seaborn

plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='Networth', data=top10, palette='viridis')
plt.title('Top 10 Billionaires by Net Worth')
plt.xlabel('Name')
plt.ylabel('Net Worth (billion USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Who had the steepest decline in net worth in a single year?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('case study - The Worlds Billionaires Dataset 1987-2022.csv')
df = df.rename(columns={'Net worth\xa0(USD)\n(billion)':'Networth'})

# Calculate year-on-year changes
df = df.sort_values(['Name', 'Year'])
df['change'] = df.groupby('Name')['Networth'].diff()
decline_row = df.loc[df['change'].idxmin()]

print(f"Billionaire with steepest decline: {decline_row['Name']} in {decline_row['Year']} with decline of ${decline_row['change']}B")

# Visualization: Line graph for that billionaire's net worth over years
person_df = df[df['Name'] == decline_row['Name']]
plt.figure(figsize=(10, 6))
plt.plot(person_df['Year'], person_df['Networth'], marker='o')
plt.title(f'Net Worth Over Time for {decline_row["Name"]}')
plt.xlabel('Year')
plt.ylabel('Net Worth (in billions)')
plt.grid(True)
plt.show()
```

**Did average billionaire wealth increase faster than inflation (approx trend by decade)?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('case study - The Worlds Billionaires Dataset 1987-2022.csv')
df = df.rename(columns={'Net worth\xa0(USD)\n(billion)':'Networth'})

# Calculate average net worth per year
average_networth_per_year = df.groupby('Year')['Networth'].mean().reset_index()

# For a simple approximation of inflation, we can assume a constant inflation rate or use a general index if available.
# Since we don't have inflation data in this dataset, we'll just show the trend of average net worth.
# To compare with inflation, we would typically need a CPI or similar index.

plt.figure(figsize=(12, 6))
plt.plot(average_networth_per_year['Year'], average_networth_per_year['Networth'], marker='o')
plt.title('Average Billionaire Net Worth Over Time')
plt.xlabel('Year')
plt.ylabel('Average Net Worth (in billions USD)')
plt.grid(True)
plt.show()

print("Average billionaire net worth per year:")
print(average_networth_per_year)
```

**Which industry dominated each decade (1980s, 1990s, 2000s, 2010s, 2020s)?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('case study - The Worlds Billionaires Dataset 1987-2022.csv')
df['decade'] = (df['Year'] // 10) * 10

# Dominant industry: most billionaires or highest total wealth? Using count of billionaires
dominant = df.groupby(['decade', 'Industry'])['Name'].count().reset_index().sort_values('Name', ascending=False).groupby('decade').head(1)

print("Dominant industry per decade:")
print(dominant)

# Visualization: Bar graph grouped by decade
pivot = df.pivot_table(index='Industry', columns='decade', values='Name', aggfunc='count', fill_value=0)
pivot.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Industry Dominance by Decade (Count of Billionaires)')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.legend(title='Decade')
plt.tight_layout()
plt.show()
```

**Correlation between age and net worth (are older billionaires richer?).**

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv('/content/case study - The Worlds Billionaires Dataset 1987-2022.csv')
df = df.rename(columns={'Net worth\xa0(USD)\n(billion)':'Networth'})

correlation = df['Age'].corr(df['Networth'])

print(f"Correlation between age and net worth: {correlation}")

# Visualization: Scatter plot with trend line
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Networth'], alpha=0.5)
# drop rows with NaN values in Age or Networth before calculating linregress
df_cleaned = df.dropna(subset=['Age', 'Networth'])
slope, intercept, r, p, std_err = stats.linregress(df_cleaned['Age'], df_cleaned['Networth'])
plt.plot(df_cleaned['Age'], slope * df_cleaned['Age'] + intercept, color='r')
plt.title('Age vs Net Worth Correlation')
plt.xlabel('Age')
plt.ylabel('Net Worth (in billions)')
plt.grid(True)
plt.show()
```

**Which companies/families appear multiple times (e.g., Walton family, Koch family)?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/case study - The Worlds Billionaires Dataset 1987-2022.csv')
df = df.rename(columns={'Net worth\xa0(USD)\n(billion)':'Networth'})
# Assume 'family' or infer from name/source, here using last name as proxy
df['family'] = df['Name'].str.split().str[-1]

family_counts = df['family'].value_counts().head(10)

print("Families/companies appearing multiple times:")
print(family_counts)

# Visualization: Bar graph
plt.figure(figsize=(10, 6))
family_counts.plot(kind='bar')
plt.title('Top Families Appearing Multiple Times')
plt.xlabel('Family Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Are there clusters of billionaires around certain industries & years?**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/case study - The Worlds Billionaires Dataset 1987-2022.csv')

# Count per industry per year
clusters = df.groupby(['Year', 'Industry'])['Name'].count().unstack()

print("Clusters (sample):")
print(clusters.head())

# Visualization: Heatmap
plt.figure(figsize=(12, 8))
plt.imshow(clusters, cmap='hot', aspect='auto')
plt.colorbar(label='Number of Billionaires')
plt.xticks(range(len(clusters.columns)), clusters.columns, rotation=90)
plt.yticks(range(len(clusters.index)), clusters.index)
plt.title('Clusters of Billionaires by Industry and Year')
plt.xlabel('Industry')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```
